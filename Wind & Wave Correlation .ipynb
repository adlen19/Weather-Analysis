# Implemented a Linear Regression Model

```python
# inputs = dff['Wind Speed ','Wind Dir']
inputs = df.filter(['Wind Speed ','Wind Dir'], axis=1)
target = dff['Wave Ht ']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(inputs, target, test_size = 0.4)
```

```python
reg = linear_model.LinearRegression()
reg.fit(X_train, y_train)   #training the model
```

```python
reg.score(X_test,y_test)
```
